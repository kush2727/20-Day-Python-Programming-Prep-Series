76)Write a program to print first non - repeating character in a string.

```python
s = input("Enter String: ")
n = len(s)
for i in range(0,n-1+1,1):
    flag = 0
    for j in range(0,n-1+1,1):
        if i!=j and s[i]==s[j]:
            flag=1
            break
    if flag == 0:
        print(s[i])
        break
```

77)write a program to check anagrams in a two strings sentence.

```python
str1 = "listen apple hello"
str2 = "silent good heart pelap"

list1 = []
word = ""
for i in str1 + " ":
    if i != " ":
        word += i
    else:
        list1.append(word)
        word = ""

list2 = []
word = ""
for i in str2 + " ":
    if i != " ":
        word += i
    else:
        list2.append(word)
        word = ""

def sort_word(w):
    chars = list(w)
    n = len(chars)


    for i in range(0,n-1+1,1):
        for j in range(0,n-2+1,1):
            if chars[j] > chars[j+1]:
                chars[j], chars[j+1] = chars[j+1], chars[j]


    sorted_word = ""
    for c in chars:
        sorted_word += c

    return sorted_word

print("Anagram word pairs:")
for i1 in list1:
    for i2 in list2:
        if sort_word(i1) == sort_word(i2):
            print(i1, "<-->", i2)
```

78)Write a program to find longest palindrome in a string sentence.

```python
sentence = "racecar belongs to malayalam actor"
words = []
word = ""

for i in sentence + " ":
    if i!= " ":
        word += i
    else:
        words.append(word)
        word = ""

def reverse_word(w):
    rev = ""
    i = len(w) - 1
    while i >= 0:
        rev += w[i]
        i -= 1
    return rev

longest = ""

for i in words:
    if i == reverse_word(i):
        if len(i) > len(longest):
            longest = i

print("Longest palindrome =", longest)
```

79) Write a program to Reverse only the vowels in a string.

```python
s = input("Enter: ")
vowels = "aeiouAEIOU"

v = ""
for ch in s:
    if ch in vowels:
        v = ch + v

result = ""
k = 0
for ch in s:
    if ch in vowels:
        result += v[k]
        k += 1
    else:
        result += ch

print(result)
```

80) Write a program to Swap characters at every even & odd position in a string.

```python
s = input("Enter: ")
n = len(s)

result = ""
i = 0

while i < n:
    if i + 1 < n:
        result += s[i+1]
        result += s[i]
        i += 2
    else:
        result += s[i]
        i += 1

print(result)
```
