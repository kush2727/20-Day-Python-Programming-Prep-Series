26) Write a program to take N inputs from the user and store them in a List.

```python
n=int(input("SIZE: "))
store=[]
for i in range(1,n+1,1):
  choose=int(input("Enter: "))
  store.append(choose)
print(store)
```

27)Write a program to Find the length of an List without using len ().

```python
a=[1,2,3,4,5,6,7,8]
count=0
for i in a:
  count+=1
print(f"Length of list: {count}")
```

28)Write a program to Find the sum of all elements in an List.

```python
ele=[20,10,30,2,3,42,12]
sum=0
for i in ele:
  sum+=i
print(f"Sum of all elements in list: {sum}")
```

29) Write a program to Print only even numbers from an List.

```python
a=[1,3,2,4,5,68,76,21,2]
even=[]
for i in a:
  if i%2==0:
    even.append(i)
print(even)
```
