86) Write a program to check if a string has a number or not without using Built-in string methods.

```python
a = input("Enter a string: ")

found = False

for i in a:
    if i >= '0' and i <= '9':
        found = True
        break

if found:
    print("The given string contain a number.")
else:
    print("The given string does not contain a number.")
```

87) Convert the string into integer, without using the built-in int() function.

```python
s = input("Enter a number string: ")

num = 0

for i in s:
    digit = ord(i) - ord('0')
    num = num * 10 + digit

print(num)
```

88) In a cricket match, the batting team is scoring runs with some run rate. The match is
for 20 overs. Write a program to find the total runs scored by the batting.
Input :13
Output: Total runs scored in 20 overs: 260.00

```python
n = int(input("Enter a number: "))

rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Number in reverse order:", rev)
```


90)Given an odd length word which should be printed from the middle of the word.




```python
s = input("Enter the word: ")

n = len(s)
mid = n // 2

result = s[mid]
print(result)

for i in range(mid + 1, n):
    result += s[i]
    print(result)

for i in range(mid - 1, -1, -1):
    result += s[i]
    print(result)
```
