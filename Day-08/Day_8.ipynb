37) Write a program to count the number of even and odd numbers in a given list.

```python
a=[22,11,28,90,45,44,22,28]
count=0
count1=0
for i in a:
  if i%2==0:
    count+=1
  else:
    count1+=1

print(f"Even Number Count: {count}")
print(f"Odd Number Count: {count1}")
```

38) Write a program to find all elements that appear more than once in a list.

```python
a=[1,1,2,3,3,4,5,6,6,7,8]
store=[]
print("Elements that appear more than once: ")
for i in a:
  if i not in store:
    store.append(i)
  elif i in store:
    print(i,end=" ")
```

39.Write a program to merge two lists and remove any duplicates from the merged list.

```python
a=[23,12,23,1,15,16,15,12]
b=[9,82,34,23,1,12,15]
c=[]
d=[]
e=[]
for i in a:
  if i not in c:
    c.append(i)
for j in b:
  if j not in d:
    d.append(j)
new_list=c+d
print(f"Merged List: {new_list}")
for k in new_list:
  if k not in e:
    e.append(k)
print(f"Merged unique List: {e}")
```
