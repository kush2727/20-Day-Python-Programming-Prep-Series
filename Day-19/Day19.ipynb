91) Given an Integer, the task is to write a program to convert integer to roman.

```python
num = int(input("Enter a number: "))

values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]

roman = ""

i = 0
while num > 0:
    if num >= values[i]:
        roman = roman + symbols[i]
        num = num - values[i]
    else:
        i = i + 1

print("Roman numeral:", roman)
```

92)Given an array of integers nums and an integer target, return indices of the two
numbers such that they add up to target.
Input: nums= [3,2,4], target= 6
Output: [1,2]

```python
nums = [3, 2,19,4,1,5]
target = 24

for i in range(0, len(nums), 1):
    for j in range(i + 1, len(nums), 1):
        if nums[i] + nums[j] == target:
            print([i, j])
            break
```

93)Write a program to check whether a given number is a Happy Number or not.

```python
n = int(input("Enter a number: "))


def sum_sq(num):
    s = 0
    while num > 0:
        d = num % 10
        s = s + d*d
        num = num // 10
    return s

list1 = []

while n != 1 and n not in list1:
    list1.append(n)
    n = sum_sq(n)

if n == 1:
    print("Happy Number")
else:
    print("Not a Happy Number")
```

94)Write a program to check whether a given number is a Harshad Number or not.


```python
n = int(input("Enter a number: "))

temp = n
sum1 = 0

while temp > 0:
    d = temp % 10
    sum1 = sum1 + d
    temp = temp // 10

if n % sum1 == 0:
    print("Harshad Number")
else:
    print("Not a Harshad Number")
```

95)Write a program to check whether a given number is a Automorphic Number or not.

```python
n = int(input("Enter a number: "))

sq = n * n

temp = n
count = 0
while temp > 0:
    count = count + 1
    temp = temp // 10

power = 1
for i in range(0, count, 1):
    power = power * 10

last = sq % power

if last == n:
    print("Automorphic Number")
else:
    print("Not Automorphic Number")
```
