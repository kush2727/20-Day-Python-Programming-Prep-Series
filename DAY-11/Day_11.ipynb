51) Write a program to move all the zeros in a list to the end while maintaining the order of the non-zero elements.

ðŸ”¹ Example:
Input: [1, 0, 2, 0, 3, 4]
Output: [1, 2, 3, 4, 0, 0]

```python
a=[12,0,2,0,34,23,12,0,134]
nonzero=[]
zero=[]
for i in a:
    if i!=0:
        nonzero.append(i)
    else:
        zero.append(i)
new=nonzero+zero
print(new)
```

52.Write a program to find the longest word in a list of strings.

```python
a = ["apple","samsung","redmi","lenovo"]
max_len = len(a[0])
for i in a:
    if len(i) > max_len:
        max_len = len(i)
longest_words = []
for i in a:
    if len(i) == max_len:
        longest_words.append(i)

print(f"The longest word(s): {longest_words}")
print(f"Length: {max_len}")
```

53.Write a program to flatten a nested list (e.g., [1,[2,3],[4,[5]]] â†’ [1,2,3,4,5])

```python
a = [1,[2],[2,3,[4,5],6],6,[7,78],324,[4,[90]]]
nested = True
while nested:
    result = []
    nested = False
    for i in a:
        if type(i) == list:
            result.extend(i)
            nested = True
        else:
            result.append(i)
    a = result

print(a)
```

54.Write a program to find the sum of all prime numbers in a list.

```python
a = [1, 2, 3, 4, 5, 7]
prime_sum = 0
for i in a:
    count = 0
    for j in range(1, i + 1, 1):
        if i % j == 0:
            count += 1
    if count == 2:
        prime_sum += i

print("Sum of all prime numbers:", prime_sum)
```
