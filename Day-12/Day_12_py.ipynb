59)Convert uppercase to lowercase and vice versa without using built-in functions.

```python
text = input("Enter a string: ")
new_text = ""

for i in text:
    val = ord(i)
    if  val>=65 and val<= 90:
        new_text += chr(val + 32)
    elif val>=97 and val<= 122:
        new_text += chr(val - 32)
    else:
        new_text += i

print("Converted string:", new_text)
```

60)Write a program to count the number of words in a string.

```python
a = "  Hello   Python  Programming  "
words = 0
in_word = False
for i in a:
    if i!= ' ' and not in_word:
        words += 1
        in_word = True
    elif i== ' ':
        in_word = False

print(words)
```
