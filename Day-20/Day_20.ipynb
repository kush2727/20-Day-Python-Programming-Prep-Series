96)﻿﻿﻿﻿Write a  program to check whether a given number is a Duck number or not.

```python
n = int(input("Enter a number: "))

def is_duck(num):
    if num <= 0:
        return False
    temp = num
    digits = 0
    while temp > 0:
        digits += 1
        temp //= 10
    temp = num
    pos = 0
    while temp > 0:
        d = temp % 10
        if d == 0 and pos != digits - 1:
            return True
        temp //= 10
        pos += 1
    return False

if is_duck(n):
    print("Duck Number")
else:
    print("Not Duck Number")
```

97)Write a program to check whether a given number is a Pronic number or not.


```python
n = int(input("Enter a number: "))

def is_pronic(num):
    k = 0

    while k * (k + 1) <= num:
        if k * (k + 1) == num:
            return True
        else:
            k += 1

    return False

if is_pronic(n):
    print("Pronic Number")
else:
    print("Not Pronic Number")

```

98)Write a program to check whether a given number is a Fascinating Number or not.



```python
n = int(input("Enter a number: "))

def is_fascinating(num):
    if num <= 0:
        return False
    a = num
    b = 2 * num
    c = 3 * num
    counts = [0] * 10
    total = 0

    def add_counts(x):
        nonlocal total
        if x == 0:
            counts[0] += 1
            total += 1
            return
        t = x
        while t > 0:
            counts[t % 10] += 1
            total += 1
            t //= 10

    add_counts(a)
    add_counts(b)
    add_counts(c)

    if total != 9:
        return False
    if counts[0] != 0:
        return False
    for d in range(1, 10):
        if counts[d] != 1:
            return False
    return True

if is_fascinating(n):
    print("Fascinating Number")
else:
    print("Not Fascinating Number")
```

99)Write a  program to check whether a given number is a Smith number or not.


```python
n = int(input("Enter a number: "))

def digit_sum(x):
    s = 0
    while x > 0:
        s += x % 10
        x //= 10
    return s

def is_prime(x):
    if x < 2:
        return False
    i = 2
    while i * i <= x:
        if x % i == 0:
            return False
        i += 1
    return True

def is_smith(num):
    if num <= 1:
        return False
    if is_prime(num):
        return False

    s_num = digit_sum(num)
    s_fact = 0
    temp = num
    f = 2

    while temp > 1:
        if temp % f == 0:
            s_fact += digit_sum(f)
            temp //= f
        else:
            f += 1

    if s_num == s_fact:
        return True
    else:
        return False

if is_smith(n):
    print("Smith Number")
else:
    print("Not Smith Number")
```

100)Write a program to check whether a given number is a Ugly number or not.

```python
n = int(input("Enter a number: "))

def is_ugly(num):
    if num <= 0:
        return False
    x = num
    while x % 2 == 0:
        x //= 2
    while x % 3 == 0:
        x //= 3
    while x % 5 == 0:
        x //= 5
    return x == 1

if is_ugly(n):
    print("Ugly Number")
else:
    print("Not Ugly Number")
```
