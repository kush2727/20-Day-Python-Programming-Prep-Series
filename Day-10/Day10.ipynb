46.Write a program to check whether a list is a palindrome (same forward and backward) without using slicing.

```python
a = [1, 1, 1, 2, 1, 1, 1]
b = []
i = 0
j = len(a) - 1
match = True

while i < j:
    if a[i] != a[j]:
        match = False
        break
    i += 1
    j -= 1

if match == True:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

 47.	Write a program to remove all occurrences of a given element from a list without using remove().

```python
a = [1, 2, 3, 4, 4, 5]
print("Original list:", a)
num = int(input("Enter a number to remove: "))
result = []
for i in a:
    if i != num:
        result.append(i)

print("Updated list:", result)
```

48.	Write a program to print the list in pairs, e.g. [1,2,3,4,5,6] â†’ (1,2) (3,4) (5,6).

```python
a = [1, 2, 3, 4, 5, 6]
pair = []
store = []
count = 0
for i in a:
    count += 1
    store.append(i)
    if count == 2:
        pair.append(tuple(store))
        store = []
        count = 0
for i in pair:
    print(i,end=" ")
```

	49.	Write a Program to print third largest number from a list.

```python
a=[10,20,13,2,1]
largest=second=third=a[0]

for i in a:
    if i>largest:
        third=second
        second=largest
        largest=i

    elif i>second and i!=largest:
        third=second
        second=i

    elif i>third and i!=second and i!=largest:
        third=i

print(third)
```
