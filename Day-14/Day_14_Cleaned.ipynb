66)Write a Python program to check whether two strings are Anagrams of each other without using built-in functions.

```python
a="silent"
b="listen"
mark=0
a1=list(a)
b1=list(b)


if len(a1)!=len(b1):
    mark=1
else:
    n=len(a1)
    for i in range(0,n-1+1,1):
        for j in range(0,n-2+1,1):
            if a1[j]>a1[j+1]:
                c=a1[j]
                a1[j]=a1[j+1]
                a1[j+1]=c


    for i in range(0,n-1+1,1):
            for j in range(0,n-2+1,1):
                if b1[j]>b1[j+1]:
                    c=b1[j]
                    b1[j]=b1[j+1]
                    b1[j+1]=c


    for i in range(0,len(a1)-1+1,1):
        if a1[i]!=b1[i]:
            mark=1


if mark==1:
    print("Not Anagram")
else:
    print("Anagram")
```

69)Write a program to reverse the order of words in a given string without using any built-in string functions like split().

```python
a = "Python is Highlevel"
word = ""
words = []
for i in a+" ":
    if i != " ":
        word += i
    else:
        words.append(word)
        word = ""
rev=""
for i in words:
    rev=i+" "+rev

print(rev)


```
