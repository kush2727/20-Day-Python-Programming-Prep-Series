63)Write a program to count vowels and consonants in a string.

```python
a="PythonProgramming"
vcount=0
count=0
for i in a:
    if i in("aeiou"):
        vcount+=1
    else:
        count+=1

print(f"vowels: {vcount}")
print(f"Consonants: {count}")
```

64)Write a program to check whether a string is a palindrome or not.

```python
a="racecar"
rev=""
for i in a:
    rev=i+rev
print(rev)

if rev==a:
    print("Palindrome")
else:
    print("Not Palindrome")
```

65)Write a program to count how many times a particular character appears in a string.

```python
a="hello"
print(a)
search=input("Enter search char: ")
count=0
for i in a:
    if i==search:
        count+=1
print(f"char count: {count}")
```
