72)Write a program to sort the string Alphabetical order.

```python
a=input("Enter String: ")
b=list(a)
n=len(a)
for i in range(0,n-1+1,1):
    for j in range(0,n-2+1,1):
        if b[j]>b[j+1]:
            c=b[j]
            b[j]=b[j+1]
            b[j+1]=c
print(b)
```

73)Write a program to remove duplicate characters from a string.


```python
a = input("Enter a string: ")
result = ""

for i in a:
    if i not in result:
        result += i

print("String after removing duplicates:", result)
```

74)Write a program to replace  all spaces with '-' from a string.

```python
a = input("Enter a string: ")
result = ""
for i in a:
    if i==" ":
        result += "-"
    else:
        result += i
print("After replacement:", result)
```

75) Write a program to find all substrings of a given string.

```python
a = input("Enter a string: ")
n = len(a)

for i in range(0, n-1+1, 1):
    sub = ""
    for j in range(i, n-1+1, 1):
        sub = sub + a[j]
        print(sub,end=" ")
```
