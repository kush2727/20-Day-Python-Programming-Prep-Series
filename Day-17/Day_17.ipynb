
81)Write a Program to Find a leap year without using modulus(%).

```python
year = int(input("Year: "))
def leap_year(year):
    if year - (year // 4) * 4 == 0:
        return True
    if year - (year // 100) * 100 == 0:
        return False
    if year - (year // 400) * 400 == 0:
        return True

if leap_year(year) == True:
    print("Leap year")
else:
    print("Not a Leap Year")
```

85)Write a Program to print prime numbers from 1 to 100 without using loops.

```python
def prime(n,i=2):
    if n<2:
        return False
    elif n==i:
        return True
    elif n%i==0:
        return False
    return prime(n,i+1)


def display(n):
    if n>100:
        return False
    elif prime(n):
        print(n,end=" ")
    display(n+1)

display(1)
```
